```python
lleve = False
if lleve:
    print('No olvides tu paraguas') 
    print('Que tengas un buen día')
    print('Hasta luego')

print ('Fin del programa')
#identación: es super importante respetarla en Python, porque define bloques de código 
#un tabulador o 4 espacios equivalen a una identación
#si no respetas la identación, el código no correrá
#el mismo VSC lo asigna automáticamente al presionar enter

#si le pongo TRUE ahí imprime todo lo que está indentado
#de lo contrario, si es falso, imprime lo que está fuera de la indentación
```

```python
lleve = True
if lleve:
    print('No olvides tu paraguas') 
    print('Que tengas un buen día')
    print('Hasta luego')

print ('Fin del programa')
```

```python
lleve = input('¿hoy llueve? (si/no): ').lower() == 'si' #asignación solo un signo =, pero si yo quiero comparar en este caso si el usuario dice si se ponene dos signos ==
if lleve:
    print('No olvides tu paraguas') 
    print('Que tengas un buen día')
    print('Hasta luego')        
print ('Fin del programa')
#si le digo que NO arriba, ejecuta y solo imprime Fin del programa
```

```python
entrada_user = input('¿hoy llueve? (si/no): ').lower() #asignación solo un signo =, pero si yo quiero comparar en este caso si el usuario dice si se ponene dos signos ==
lleve =  entrada_user == 'si' #si abajo ponemos if entrada_usar funcionará igual, pero a veces el texto es muy largo aquí, entonces conviene convertir en la variable lleve
if lleve:
    print('No olvides tu paraguas') 
    print('Que tengas un buen día')
    print('Hasta luego')

print ('Fin del programa')

#si le digo que NO arriba, ejecuta y solo imprime Fin del programa
#si ejecuto y pongo si, imprime todo lo de adentro del if y luego Fin del programa

#el LOWER sirve para que sirva mayuscula y minuscula, porque eso lo convierte todo a minuscula

#a esto le podríamos agregar muchas mas cosas, por ejemplo, que solo pueda poner si o no
#o que si pone otra palabra, entonces le diga que ponga si o no
```
